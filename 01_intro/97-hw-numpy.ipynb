10) Создать матрицу случайных чисел размера 10 на 10 и отнять среднее по строкам из каждого элемента строки

```python
random_matrix = np.random.rand(10, 10)
row_means = np.mean(random_matrix, axis=1)
normalized_matrix = random_matrix - row_means[:, np.newaxis]
print("Оригинальная матрица:")
print(random_matrix)
print("Средние по строкам:")
print(row_means)
print("Нормализованная матрица:")
print(normalized_matrix)
```

11) Создать матрицу случайных чисел размера 10 на 10 и отнять медиану по столбцам из каждого элемента столбца  

```python
random_matrix = np.random.rand(10, 10)
column_medians = np.median(random_matrix, axis=0)
normalized_matrix = random_matrix - column_medians
print("Оригинальная матрица:")
print(random_matrix)
print("Медианы по столбцам:")
print(column_medians)
print("Нормализованная матрица:")
print(normalized_matrix)
```

13) Вектор чисел от 1 до 51 преобразовать в вектор нечетных чисел и в вектор четных чисел, поменять их размерности, чтобы получились две матрица. Найти произведение этих матриц. Найти обратную матрицу этого произведения.  

```python
import numpy as np

vector = np.arange(1, 52)

odd_numbers = vector[vector % 2 != 0]
even_numbers = vector[vector % 2 == 0]
odd_matrix = odd_numbers[:25].reshape(5, 5)
even_matrix = even_numbers[:25].reshape(5, 5)

product_matrix = np.dot(odd_matrix, even_matrix)

if np.linalg.det(product_matrix) != 0:
    inverse_matrix = np.linalg.inv(product_matrix)
    print("Обратная матрица:", inverse_matrix)
else:
    print("Обратная матрица не существует, так как определитель равен 0.")

```

14) Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.
